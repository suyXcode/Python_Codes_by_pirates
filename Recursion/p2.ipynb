```python
#WAP to print sum of numbers from 1 to 100 by using recursion.

def sum(n):
    if n==1:
        return 1
    else:
        return n+sum(n-1)
n=int(input("Enter a number: "))
print("Sum of numbers from 1 to", n, "is:", sum(n))
```

```python
#WAP to print sum of squares of numbers from 1 to 100 by using recursion.
def sum_of_sq(n):
    if n==1:
        return 1
    else:
        return n * n + sum_of_sq(n-1)
n=int(input("Enter a number: "))
print("Sum of squares of numbers from 1 to", n, "is:", sum_of_sq(n))
```

```python
#WAP to print sum of cubes of numbers from 1 to 100 by using recursion.
def sum_of_cube(n):
    if n==1:
        return 1
    else:
        return n * n * n + sum_of_cube(n-1)
n=int(input("Enter a number: "))
print("Sum of cubes of numbers from 1 to", n, "is:", sum_of_cube(n))
```

```python
#WAP to calculate factorial by using recursion.
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
num=int(input("Enter a number to find its factorial: "))
print(f"The factorial of {num} is {factorial(num)}")
```

```python
# WAP to calculate power using recursion
def power(n, p):
    if p == 0:
        return 1
    else:
        return n * power(n, p - 1)

num = int(input("Enter the number: "))
exp = int(input("Enter the power: "))

print(f"{num} raised to the power of {exp} is {power(num, exp)}")
```

```python
#WAP to get digit count of a number by using recursion.
def digit_count(n):
    if n==0:
        return 0
    else:
        return 1+digit_count(n//10)
num=int(input("Enter a number to find its digit count: "))
print(f"The digit count of {num} is {digit_count(num)}")
```

```python
#WAP  to get digit sum of a number by using recursion.
def digit_sum(n):
    if n==0:
        return 0
    else:
        return n%10 + digit_sum(n//10)
num=int(input("Enter a number to find its digit sum: "))
print(f"The digit sum of {num} is {digit_sum(num)}")
```

```python
#WAP to reverse a number by using recursion.
def reverse_num(n,rev=0):
    if n==0:
        return rev
    else:
        return reverse_num(n//10, rev*10 + n%10)

num=int(input("Enter a number to reverse: "))
print(f"The reverse of {num} is {reverse_num(num)}")
```

```python
#WAP to check  a palindrome number by using recursion.
def is_palindrome(n, rev=0):
    if n == 0:
        return rev
    else:
        return is_palindrome(n//10, rev*10 + n%10)

num = int(input("Enter a number to check if it's a palindrome: "))
if num == is_palindrome(num):
    print(f"{num} is a palindrome.")
else:
    print(f"{num} is not a palindrome.")
```
