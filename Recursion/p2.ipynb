```python
# WAP to calculate nth term of Fibonacci series using recursion
def fibonacci(n):
    if n <= 0:
        return "Invalid"
    elif n == 1:
        return 0     
    elif n == 2:
        return 1    
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Example use
n = int(input("Enter the term number (n): "))
print(f"The {n}th term of the Fibonacci series is: {fibonacci(n)}")
```

```python
# WAP to print all prime numbers up to n using recursion
def is_prime(num, i=2):
    # Base cases
    if num <= 1:
        return False
    if i * i > num:
        return True
    if num % i == 0:
        return False
    return is_prime(num, i + 1)

def print_primes(n, current=2):
    if current > n:
        return
    if is_prime(current):
        print(current, end=" ")
    print_primes(n, current + 1)

# Example use
n = int(input("Enter a number: "))
print(f"Prime numbers up to {n} are:")
print_primes(n)
```

```python
#18. 342. Power of Four
# Given an integer n, return true if it is a power of four. Otherwise, return false.

# An integer n is a power of four, if there exists an integer x such that n == 4x.

 

# Example 1:

# Input: n = 16
# Output: true
# Example 2:

# Input: n = 5
# Output: false
# Example 3:

# Input: n = 1
# Output: true

# Constraints:

# -231 <= n <= 231 - 1


def is_power_of_four(n):
    if n <= 0:
        return False
    elif n == 1:
        return True
    elif n % 4 != 0:
        return False
    else:
        return is_power_of_four(n // 4)

print(is_power_of_four(16))  
print(is_power_of_four(128))   
```

```python
#19 WAP to take three user inputs and print HCF of the three numbers.
def findGCD(num1, num2):
    if num1 == 0 or num2 == 0:
        return num1 + num2
    # base case
    if num1 == num2:
        return num1
    
    # num1>num2
    if num1 > num2:
        return findGCD(num1 - num2, num2)
    else:
        return findGCD(num1, num2 - num1)

def hcf_of_three(a, b, c):
    return findGCD(findGCD(a, b), c)
print(hcf_of_three(48, 64, 16))

```

```python
# WAP to take three user inputs and print LCM of the three numbers.
def lcm(a, b):
    return (a * b) // findGCD(a, b)

def lcm_of_three(a, b, c):
    return lcm(lcm(a, b), c)

print(lcm_of_three(4, 5, 6))
```
