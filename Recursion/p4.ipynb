```python
def fact(n):
    if n==0:
        return 1
    if n==1:
        return 1
    return n*fact(n-1)

print(fact(0))
```

```python
def power(b,p):
    if p == 0:
        return 1
    return b * power(b,p-1)

print(power(2,3))
```

```python
def count(n):
    if n<10:
        return 1
    if n==0:
        return 0
    
    return 1+count(n//10)

print(count(1234))
```

```python
def reverse(n,rev=0):
    if n==0:
        return rev
    return reverse(n//10,rev*10 + n%10)

print(reverse(123))
```

```python
def is_p(n):
    if str(n)[0]!=str(n)[-1]:
        return False
    if len(str(n))<=1:
        return True
    return is_p(str(n)[1:-1])

print(is_p(12321))
```

```python
def fib(n):
    if n<=0:
        return "invalid"
    if n==1:
        return 0
    if n==2:
        return 1
    return fib(n-1)+fib(n-2)

print(fib(3))
```

```python
def isPrime(num,i=2):
    if num <= 1:
        return False
    if i*i > num:
        return True
    if num%i==0:
        return False
    return isPrime(num,i+1)
print(isPrime(13))
```

```python
def printP(n,c=2):
    if c>n:
        return 
    if isPrime(c):
        print(c,end=" ")
    printP(n,c+1)
printP(12)
```

```python
def ispowertwo(n):
    if n<0:
        return False
    
    elif n==1:
        return True
    elif n%2!=0:
        return False
    return ispowertwo(n//2)

print(ispowertwo(128))
```

```python
def ispowerthree(n):
    if n<0:
        return False
    elif n==1:
        return True
    elif n%3!=0:
        return False
    return ispowerthree(n//3)

print(ispowerthree(66))
```

```python
def hcf(a,b):
    if b==0:
        return a
    return hcf(b,a%b)
print(hcf(16,24))
```

```python
def h3(a,b,c):
    return hcf(hcf(a,b),c)
print(h3(4,5,6))
```

```python
def lcm(a,b):
    return (a*b)// hcf(a,b)
print(lcm(24,3))
```

```python
def l3(a,b,c):
    return lcm(lcm(a,b),c)
print(l3(2,3,5))
```

```python
def toh(n,s,t,a):
    if n==1:
        print(f"move disk 1 from {s} to {t}")
        return
    toh(n-1,s,a,t)
    print(f"Move disk {n} from {s} to {t}")
    toh(n-1,a,t,s)

n=3
toh(n,"A","B","C")
```

```python
def is_ugly(n):
    if n<=0:
        return False
    while n%2==0:
        n//=2
    while n%3==0:
        n//=3
    while n%5==0:
        n//=5
    
    return n == 1

print(is_ugly(13))
```

```python
def is_ugly(n):
    if n<=0:
        return False
    if n==1:
        return True
    elif n%2 == 0:
        return is_ugly(n//2)
    elif n%3 == 0:
        return is_ugly(n//3)
    elif n%5 == 0:
        return is_ugly(n//5)
    return False

print(is_ugly(14))
```

```python
import time
def isu(n):
    start = time.time()
    if n <= 0:
        print("Not an Ugly Number")
    else:
        for i in [2, 3, 5]:
            while n % i == 0:
                n //= i
        if n == 1:
            print("It is an Ugly Number")
        else:
            print("Not an Ugly Number")
    end = time.time()
    print("Time taken:", end - start)
isu(14)
```

```python
def mirror(n):
    if n==0:
        return ""
    if n==1:
        return "1"
    return str(n) + mirror(n-1) + str(n)

res = mirror(5)
print(res)
```
