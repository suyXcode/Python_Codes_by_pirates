```python
#WAJP to print sum of numbers from 1 to 100 by using recursion.

def sum_numbers(n):
    if n == 1:
        return 1
    else:
        return n + sum_numbers(n - 1)
print(sum_numbers(100))  # Output: 5050
```

```python
#WAP to print sum of squares of numbers from 1 to 100 by using recursion.

def sum_of_squares(n):
    if n == 1:
        return 1
    else:
        return n * n + sum_of_squares(n - 1)
print(sum_of_squares(100))  # Output: 338350
```

```python
#WAP to print sum of cubes of numbers from 1 to 100 by using recursion.

def sum_of_cubes(n):
    if n == 1:
        return 1
    else:
        return n * n * n + sum_of_cubes(n - 1)

print(sum_of_cubes(100))  
```

```python
#WAJP to calculate factorial by using recursion.

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(5))  
```

```python
#WAJP to calculate power by using recursion.
def power(base, exp):
    if exp == 0:
        return 1
    else:
        return base * power(base, exp - 1)
print(power(2, 3))  # Output: 8
```

```python
#WAJP to get digit count of a number by using recursion.

def count_digits(n):
    if n == 0:
        return 0 
    else:
        return 1 + count_digits(n // 10)

print(count_digits(12345))  # Output: 5
```

```python
#WAJP to get digit sum of a number by using recursion.

def sum_of_digits(n):
    if n == 0:
        return 0
    else:
        return n % 10 + sum_of_digits(n // 10)

print(sum_of_digits(12345))  # Output: 15
```

```python
#WAP to reverse a number by using recursion.
def reverse_number(n, rev=0):
    if n == 0:
        return rev
    else:
        rev = rev * 10 + n % 10
        return reverse_number(n // 10, rev)
print(reverse_number(1234))  # Output: 4321
```

```python
#WAP to check  a palindrome number by using recursion.

def is_palindrome(n, rev=0, original=None):
    if original is None:
        original = n
    if n == 0:
        return rev == original
    else:
        rev = rev * 10 + n % 10
        return is_palindrome(n // 10, rev, original)
print(is_palindrome(121))  # Output: True
print(is_palindrome(123))  # Output: False
```

```python
#WAJP to calculate nth term of Fibonacci series by using recursion.

def fib(n):
    if n < 2:
        return n
    
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

print(fib(7))  # Output: 13

```

```python
#16 Given an integer n, return true if it is a power of two. Otherwise, return false.

# An integer n is a power of two, if there exists an integer x such that n == 2x.
# Example 1:

# Input: n = 1
# Output: true
# Explanation: 20 = 1
# Example 2:

# Input: n = 16
# Output: true
# Explanation: 24 = 16
# Example 3:

# Input: n = 3
# Output: false

def is_power_of_two(n):
    if n <= 0:
        return False
    elif n == 1:
        return True
    else:
        return is_power_of_two(n // 2)
print(is_power_of_two(16))  
print(is_power_of_two(32))   
```

```python
# 342. Power of Four
# Given an integer n, return true if it is a power of four. Otherwise, return false.

# An integer n is a power of four, if there exists an integer x such that n == 4x.

 

# Example 1:

# Input: n = 16
# Output: true
# Example 2:

# Input: n = 5
# Output: false
# Example 3:

# Input: n = 1
# Output: true
 

# Constraints:

# -231 <= n <= 231 - 1
def is_power_of_four(n):
    if n <= 0:
        return False
    elif n == 1:
        return True
    elif n % 4 != 0:
        return False
    else:
        return is_power_of_four(n / 4)
print(is_power_of_four(16))
print(is_power_of_four(81))
```

```python
#19 WAJP to take three user inputs and print HCF of the three numbers.
def hcf(a, b):
    if b == 0:
        return a
    else:
        return hcf(b, a % b)

def hcf_of_three(a, b, c):
    return hcf(hcf(a, b), c)
print(hcf_of_three(48, 64, 16))  # Output: 16   
```

```python
# WAJP to take three user inputs and print LCM of the three numbers.

def lcm(a, b):
    return (a * b) // hcf(a, b)

def lcm_of_three(a, b, c):
    return lcm(lcm(a, b), c)

print(lcm_of_three(4, 5, 6))  # Output: 60
```

```python
#WAP to solve Tower Of Hanoi problem.

def tower_of_hanoi(n, source, target, auxiliary):
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return
    tower_of_hanoi(n - 1, source, auxiliary, target)
    print(f"Move disk {n} from {source} to {target}")
    tower_of_hanoi(n - 1, auxiliary, target, source)
    
# Example usage:
n = 3  # Number of disks
tower_of_hanoi(n, 'A', 'C', 'B')  # A, B and C are names of rods
```

```python
# #22. Ugly Number
# #263. Ugly Number
# Easy
# Topics
# premium lock icon
# Companies
# An ugly number is a positive integer which does not have a prime factor other than 2, 3, and 5.

# Given an integer n, return true if n is an ugly number.

# Example 1:

# Input: n = 6
# Output: true
# Explanation: 6 = 2 Ã— 3
# Example 2:

# Input: n = 1
# Output: true
# Explanation: 1 has no prime factors.
# Example 3:

# Input: n = 14
# Output: false
# Explanation: 14 is not ugly since it includes the prime factor 7.
# Constraints:
# -231 <= n <= 231 - 1

def is_ugly(n):
    if n <= 0:
        return False
    for factor in [2, 3, 5]:
        while n % factor == 0:
            n //= factor
    return n == 1

print(is_ugly(6))   
print(is_ugly(14))  
```
