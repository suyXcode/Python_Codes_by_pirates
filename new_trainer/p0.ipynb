# types of method üëç

* class method : 
-- It is used to access and modify class members.
-- While declaring class method we should pass class decorator. @classmethod
-- pass first parameter (cls)

* Object Method : 
-- It is used to access and modify object members.
-- We will pass first argument self.
-- Not use decorator

* Static Method üëç
-- It does not belong any class method and object method, But it acts as helper method.
-- while declaring static method we will pass first @.





```python
class Company:
    cname= "Qspider"
    cloc = "Noida"
    chead = "Shambhu Sir"
    def __init__(self, ename, eid,email,role):
        self.ename =ename 
        self.eid = eid 
        self.email = email
        self.role = role
    def display(self):
        print("---------------------------------------------")
        print("Company Details :")
        print(" ")
        print("Company Name :",Company.cname)
        print("Company Location :",Company.cloc)
        print("Company Head :",Company.chead)
        print("---------------------------------------------")
        print("Employee Details :")
        print(" ")
        print("Employee Name :",self.ename)
        print("Employee ID :",self.eid)
        print("Employee Email :",self.email)
        print("Employee Role :",self.role)
        print("---------------------------------------------")

emp1 = Company("Ankit",12,"Ankit@gmail.com","Project_Manager")
emp2 = Company("Anuj",13,"Anuj@gmail.com","Project_Leader")
emp3 = Company("Suraj",15,"Suraj@gmail.com","Senior Developer")
emp4 = Company("Shaad",16,"Shaad@gmail.com","Project_Head")
emp5 = Company("Shekhar",19,"Shekhar@gmail.com","Mobile Developer")
emp6 = Company("Shiv",18,"Shiv@gmail.com","Jr. Developer")

```

```python
# create a class Bank with 3 class properties and 3 object members with 3 object properties  with modification

class Bank:
    bname="Sbi"
    loc="Noida"
    bhead="Ankit"
    website = "sbi.com"
    helpline=123456789

    def __init__(self,name,age,balance):
        self.name=name
        self.age=age
        self.balance=balance
    
    def ch_name(self,new_name):
        self.name=new_name
    
    def ch_age(self,new_age):
        self.age=new_age
    
    def ch_balance(self,new_balance):
        self.balance=new_balance
    
    
    @classmethod
    def ch_bhead(cls,new_name):
        cls.bhead = new_name
    @classmethod
    def ch_loc(cls,new_loc):
        cls.loc = new_loc
    
    
    
    
    def show(self):
        print("---------------------------------------------")
        print("Bank Details :")
        print(" ")
        print("Bank Name :",Bank.bname)
        print("Bank Location :",Bank.loc)
        print("Bank Head :",Bank.bhead)
        print("Bank website :",Bank.website)
        print("Bank Helpline :",Bank.helpline)
        print("---------------------------------------------")
        print("Customer Details :")
        print(" ")
        print("Customer Name :",self.name)
        print("Customer Age :",self.age)
        print("Customer Balance :",self.balance)
        print("---------------------------------------------")

cus1=Bank("Suraj",22,3500000)
cus2=Bank("Ankit",22,3500000)
cus3=Bank("Anuj",22,3500000)
```
