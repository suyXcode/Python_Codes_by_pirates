```python
# 26. WAP to check whether a number is Xylem or Phloem (using for loop)

num = int(input("Enter the number: "))
s = str(num)

first = int(s[0])         
last = int(s[-1])         
mean_sum = 0

for i in range(1, len(s)-1):
    mean_sum += int(s[i])

ex_sum = first + last

if mean_sum == ex_sum:
    print(num, "is a Xylem Number.")
else:
    print(num, "is a Phloem Number.")
```

**Intermediate Termination**

These are statements which is used to terminate   block  of the ongoing loop.
break
continue 
pass

These are used inside of a loop.



**break :**
It is keyword which is used to terminate the flow of looping statement.
It is written with a condition. When the condition become true its terminate the loop.
It will not execute any statement just break keyword.


Syntax:

for var in collection:
    if condition:
        break




**continue:**
It is a keyword ,used to skip current itration.It returns the flow back to the loop.

syntax:

for i in range(7):
    if i==3:
        continue
    print(i,end=" ")  #Output:-   1 2 4 5 6


**pass üëç:**

It is used to make a empty block as a valid block.
It can used inside of loop,function,method,class etc.
It is used to keep function , class ,method resevered in future use. 

Syntax : 
for i in range(7):
    pass
