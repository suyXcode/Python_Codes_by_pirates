# Comprehension:
* A conscie way of creating a new mutable collection.

* conscie : single expression /single line of statement.
  

There are three type of comprehension.
* list comprehension
* set comprehension
* dictionary comprehension


List Comprehension ðŸ‘
* var=[var for var in collection]   --> simple looping statement
* var=[var for var in collection if <condition>]  --> 1 statement block
* var=[TSB if condition else FSB for var in collection]     --> 2 statement block
* var = [val fro var1 in condition for var2 in collection]   ---> nested loop

Set Comprehension ðŸ‘
* var={var for var in collection}   --> simple looping statement
* var={var for var in collection if <condition>}  --> 1 statement block
* var={TSB if condition else FSB for var in collection}     --> 2 statement block
* var = {val fro var1 in condition for var2 in collection}   ---> nested loop




```python
#wap list comprehension to print n natural number.
n=int(input("Enter the num :"))
l=[i for i in range(1,n+1)]
print(l)
```

```python
#wap to create a list odd
n=int(input("Enter the num :"))
l=[i for i in range(1,n+1,2)]
print(l)
```

```python
#wap to create a list odd
n=int(input("Enter the num :"))
l=[i for i in range(1,n+1) if i%2==1]
print(l)
```

```python
#wap to create a list of even number
n=int(input("Enter the num :"))
l=[i**2 if i%2==0 else i**3 for i in range(1,n+1)]
print(l)
```

Set comprehension

```python
# t1=(10,"hello",5,3-2j,11,'java',[1,2,3],10,5,11)

# set1 = [i[::-1] if type(i) == [list,tuple,dict,set ] or type(i)== str else i for i in t1]
# print(set1)

t1 = (10, "hello", 5, 3-2j, 11, 'java', (1,2,3), 10, 5, 11)

set1 = {
    i[::-1] if type(i)==str else i
    for i in t1
}

print(set1)
```

```python
n=int(input("Enter the num"))
set1 ={i for i in range(1,n+1) if i%2==1}
print(set1)
```

# Dictionary Comprehension

Syntax

* var = {key:value for var in collection}
* var = {key:value for var in collection if condition}
* var ={key:value if condition else val2 for var in collection}


```python
# wap to create a dictionary of n natural number such that the value should be square of the number.
n= int((input("Enter the num")))
var ={i:i**2 for i in range(1,n+1)}
print(var)
```

```python
#Create a dictionary of n natural number such that if the number is odd the value should be cube of the number and if the number is even the value should be square of the number.
#Input : 5
l1=['hello','world',55,1-3j,22.5,'java','python']
l2={i:i[::-1] for i in l1 if type(i)==str}
print(l2)
```

```python
#Create a dictionary of n natural number such that if the number is odd the value should be cube of the number and if the number is even the value should be square of the number.
#Input : 5
n=int(input("Enter the num :"))
d={i:i**3 if i%2==1 else i**2 for i in range(1,n+1)}
print(d)
```
