```python
# Python‚Äôs built-in list.remove() ‚Äî it removes the first occurrence of a given element from the list.
def my_remove(lst, value):
    for i in range(len(lst)):
        if lst[i] == value:
            del lst[i]
            return  

nums = [10, 20, 30, 20, 40]
print("Before:", nums)

my_remove(nums, 20)
print("After removing 20:", nums)

my_remove(nums, 100) 
```

```python
def rem(l, a):
    if type(l) not in (list, set):
        print("Not Applicable")
        return

    li = []
    flag = False
    for i in l:
        if i == a and not flag:
            flag = True
            continue
        li.append(i)
    print(li)

# Example
rem([1, 2, 2, 4], 2)
```

```python
def is_prime(num):
    for i in range(2,num):
        if num%i==0:
            return False
    else:
        return True
n= int(input("Enter the numbe :"))
print("Number is prime  :",is_prime(n))

```
