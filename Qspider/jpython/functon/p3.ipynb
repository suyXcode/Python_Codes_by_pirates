```python
# write a function to check if a number is armstrong or not
def is_armstrong(num):
    str_num = str(num)
    num_digits = len(str_num)
    sum_of_cubes = sum(int(digit) ** num_digits for digit in str_num)
    return sum_of_cubes == num

print("153 is armstrong:",is_armstrong(153))  
print("9474 is armstrong:",is_armstrong(9474)) 
```

```python
# armstrong number series 
n= int(input("Enter the number :"))
count =0
for i in range(1,n+1):
    if is_armstrong(i):
        print(i,end=" ")
        count += 1
print("\nTotal armstrong numbers between 1 and", n, "are:", count)
```

```python
# WAF to get the sum of digits of a number until the sum becomes a single digit

def sum_s(num):
    while num >= 10:        
        total = 0
        for digit in str(num):
            total += int(digit)
        num = total          

    return num

print("sum of digits of a number",f"({1199})" ,"until the sum becomes a single digit is ",sum_s(1199))
```

# Argument : 
------------

These are the values pass to the function for further processing.


**Formal argument üëç (Parameters)**
--------------------------------

-- known as Parameters, who is receive the values during the function call.
-- pass in function defintion.


**Actual argument :(Actual Values)**
----------------------------------

-- These are the actual values passed to function 
-- pass in function calling.

Syntax: 
--------

def fname(parameter):
    sb 
fname(argument)
    

**Types of Argument :**
--------------------

There are five types of arguments.

1. Positional Argument
2. Keyword Argument
3. Default Argument
4. Variable Length Argument
5. Variable Length Keyword Argument


**Positional Argument** 
----------------------------

-- These are present function declaration / definition
-- the length of formal argument should be equal to actual arguments.
-- here order and position are matters


Syntax : 
--------

def f(argu1,argu2,...,argun):
    sb
f(val1,val2,...,valn)

Example:

def f(a,b,c):
    print(a,b,c)
f(1,23,4)


**Keyword Argument**
---------------------

--- Pass in function call
--- these are pass in form of key and value
--- the length of formal argument should be equal to actual arguments.
--- Here order and position not matter

**Syntax**
----------
def f(a1,a2,...,an):
    sb 
f(key1=val1,key2=val2,.......,keyn=valn)


**Example**
-------------

def f(a,b,c):
    print(a,b,c)
    print(c,a,b)
f(a=1,c=2,b=3)



**Default Arguments**
---------------------

--- These are present in definition
--- passed in form of key and value
--- here the length of actual argument  may be equal or less than the length of formal argument.
--- order and position matters
--- if the user will not provide any values then the default values are considered.


**Syntax**
def f(key=value,key=value):
    sb 
f(val1,val2)



**Example**

def main(a=0,b=0,c=0):
    print(a,b,c)
main(10,20)


Note : if we using both positional argu and keyword argu then keyword argu should followed positional argu.


**Variable Length Argument**
----------------------------

--- It is used to n number of argument 
--- it is denoted *args
--- We should use arg*
--- It store n number of value in the form of tuple.





**Example**
------------

def main(*args):
    print(args)
main(10,20,"hi",True)




**Variable Length Keyword Arguments**
--------------------------------------

--- It is used to receive n number of keyword argument 
--- it is defined in parameter  **kwargs
--- It store n number of value in the form of dictionary.
 


**Syntax**:
-----------

def f(**kwargs):
    sb 
f(key=value,key=values)



**Example**
------------

def main(**kwargs):
    print(kwargs)
main(a=10,b=20,c="hi",d=True)

-- It is process of combining/binding or packing multiple values in a single variable .It is used during the transfer of data especially in function call.

1. Single Packing (in form of tuple)
2. Double Packing (in form of Dict)

Apart from dict and tuple we can pass value in list. 




**Unpacking**
-- It is proccess of extracting value from packed structure into seperate variable.
