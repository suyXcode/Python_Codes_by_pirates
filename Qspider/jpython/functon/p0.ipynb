```python
# Pythons built-in list.remove()  it removes the first occurrence of a given element from the list.
def my_remove(lst, value):
    for i in range(len(lst)):
        if lst[i] == value:
            del lst[i]
            return  # Exit after removing the first match
    # else :
    #     print("ValueError: list.remove(x): x not in list")  # If value not found


# Example use
nums = [10, 20, 30, 20, 40]
print("Before:", nums)

my_remove(nums, 20)
print("After removing 20:", nums)

my_remove(nums, 100)  # Trying to remove a value not in list
```

```python

def rem(l, a):
    if type(l) not in (list, set):
        print("Not Applicable")
        return

    li = []
    found = False
    for i in l:
        if i == a and not found:
            found = True
            continue
        li.append(i)
    print(li)

# Example
rem([1, 2, 2, 4], 2)
```

```python
#implementation of Pythons index() function 
#(It returns first index of a value,if not found prints error message)

def my_index(lst, value):
    if type(lst) not in (list, tuple, str):
        print("Not Applicable")
        return
    
    pos = -1
    for i in range(len(lst)):
        if lst[i] == value:
            pos = i
            break   # stop after first match
    
    if pos == -1:
        print("ValueError: value not found")
    else:
        print(pos)

# Example
my_index([10, 20, 30, 20], 20)
```
